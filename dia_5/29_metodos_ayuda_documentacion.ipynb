```python
#método insert() introduce o inserta en una lista o conjunto mutable, en el índice elegido, el elemento indicado entre comillas
frutas = ["mango", "banana", "cereza", "ciruela", "pomelo"]
frutas.insert(3, 'manzana')
frutas
```

```python
#método isdisjoint() se utiliza para comparar conjuntos. Su sintaxis es: CONJUNTO1.isdisjoint(CONJUNTO2)
marcas_smartphones = {"Samsung", "Xiaomi", "Apple", "Huawei", "LG"}

marcas_tv = {"Sony", "Philips", "Samsung", "LG"}

conjuntos_aislados = marcas_smartphones.isdisjoint(marcas_tv)
conjuntos_aislados
```
